```python
start = int(input("Enter the starting number: "))
end = int(input("Enter the ending number: "))
loop_counter = 0
while start <= end:
    if start % 2 == 0:
      print(start)
    start += 1
    loop_counter += 1
print(f"Loop executed {loop_counter} times.")
```

```python
start = int(input("Enter the starting number: "))
end = int(input("Enter the ending number: "))   
if start%2 ==0:
   i = start
   print(i)
else:
     i = start + 1
     print(i)
while i <= end:
    print(i)
    i += 2
print(f"Loop executed {(end - start) // 2 + 1} times.") 
```

```python
x = 1 
while x <= 100:
    if x % 3 == 0 and x % 5 == 0:
        print("FizzBuzz")
    elif x % 3 == 0:
        print("Fizz")
    elif x % 5 == 0:
        print("Buzz")
    else:
        print(x, end='\t')
    x += 1
```

```python
no_of_times = 3
password = "python"
atempts = 0
while atempts < no_of_times:
    user_input = input("Enter the password: ")
    if user_input == password:
        print("Access granted!")
        break
    else:
        print("Incorrect password",(no_of_times - atempts-1 ), "attempts left.")
    atempts += 1
if atempts == no_of_times:
    print("Access denied! Too many attempts.")
```
